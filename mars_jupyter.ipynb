```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
import pymongo
import numpy as np
from pprint import pprint
```

```python
file_url = 'https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest'


page = requests.get(file_url)
print(page.status_code)
# print(page.content)

soup=bs(page.content,'html.parser')
print(soup.prettify())
```

```python
# soup.find('ul',{"class":"item_list"}).find_all('li')[0].find('a')
title = soup.find_all('div', class_="content_title")
# for link in tb.find_all('li'):
#     name=link.find('a')
print(title)
```

```python
title_list = []
for result in title:
    # Error handling
    try:
        # Identify and return title of listing
        title = result.find('a').text
        print(title)
        title_list.append(title)
    except AttributeError as e:
        print(e)
print(len(title_list))
```

```python
paragraph = soup.find_all('div',class_= 'rollover_description')
print(paragraph)
```

```python
paragraph_list=[]

for result in paragraph:
    # Error handling
    try:
        text = result.text
        paragraph_list.append(text)
        print(text)
    except AttributeError as e:
        print(e)
print(len(paragraph_list))
```
